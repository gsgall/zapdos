```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.colorbar import ColorbarBase
from matplotlib.colors import Normalize
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
# plt.rcParams['figure.dpi'] = 600
font = 15
plt.rcParams['axes.titlesize'] = font
plt.rcParams['axes.labelsize'] = font

plt.rcParams['xtick.labelsize'] = font-1
plt.rcParams['ytick.labelsize'] = font-1
```

```python

# Create a figure and axis
fig, ax = plt.subplots()

im = mpimg.imread('8mm_base.eps')

ax.set_ylim(-1.7, 8)
ax.set_xlim(0, 12)
ax.set_xticks([0, 5, 10])
ax.set_yticks([0, 2, 4, 6, 8])
plt.gca().invert_yaxis()

ax.set_xlabel('X [mm]')

ax.set_ylabel('Dist. From Nozzle [mm]', labelpad=10)
ax.set_zorder(2)
ax.set_facecolor('none')
ax.set_title('He into N$_2$ atmosphere')

ax_tw_x = ax.twinx()
ax_tw_x.axis('off')
ax2 = ax_tw_x.twiny()
ax2.imshow(im, aspect='auto')
ax2.axis('off')

cmap = plt.cm.jet  # You can choose a different colormap
norm = Normalize(vmin=0, vmax=1)
# Add a color bar
cax = ColorbarBase(ax=fig.add_axes([0.95, 0.11, 0.02, 0.78]), cmap=cmap, norm=norm, orientation='vertical')
cax.set_label('w$_{\mathrm{He}}$')

plt.savefig('8mm.eps', format='eps', bbox_inches='tight')
plt.show()
```
