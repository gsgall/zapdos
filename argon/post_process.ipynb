```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./argon_out.csv")

period_end = 1 / 13.56e6
print(period_end * 14000)
```

```python
length = 30e-3
depth = 1e-3
end_idx = 180
periods = 14000
plt.figure()
for i in range(periods):
  plt.axvline(i * 1 / 13.56e6 + 3.7e-9, linestyle="dashed", c="k")

plt.plot(data["time"], data["periodic_power"] * 13.56e6 * length * depth)

# plt.xlim(left=9.99e-4, right=1e-3)
plt.xlabel("Time [s]")
plt.ylabel("Average Power Deposition [W]")
plt.show()
```
